Numpy

```python
import pandas as pd
import numpy as np

```

```python
df = pd.read_csv("1774404110.csv")
```

```python
df.head()
```

1

```python
mpg = df['mpg'].to_numpy()

mean_mpg = np.mean(mpg)
median_mpg = np.median(mpg)
std_mpg = np.std(mpg)

print("Mean:", mean_mpg)
print("Median:", median_mpg)
print("Std Dev:", std_mpg)

# Number of cars with mpg > 25
count_above_25 = np.sum(mpg > 25)
print("Cars with mpg > 25:", count_above_25)

```

2


```python
cylinders = df['cylinders'].to_numpy()
car_names = df['car name'].to_numpy()


filtered_cars = car_names[cylinders > 6]


filtered_cars_list = filtered_cars.tolist()

print(filtered_cars_list)
```

3

```python
horsepower = df['horsepower'].to_numpy()

# Replace '?' with NaN and then convert to float
horsepower[horsepower == '?'] = np.nan
horsepower = horsepower.astype(float)

mean_hp = np.nanmean(horsepower)
horsepower[np.isnan(horsepower)] = mean_hp

updated_hp = horsepower * 1.10

print(updated_hp)
```

4

```python
displacement = df['displacement'].to_numpy()
origin = df['origin'].to_numpy()


europe_disp = displacement[origin == 2]


avg_disp_europe = np.mean(europe_disp)

print(avg_disp_europe)
```

5

```python
cylinders = df['cylinders'].to_numpy()
mpg = df['mpg'].to_numpy()

unique_cyl = np.unique(cylinders)

for cyl in unique_cyl:
    mean_val = np.mean(mpg[cylinders == cyl])
    print(f"Cylinders {cyl}: Mean MPG = {mean_val}")
```

```python
import pandas as pd
```

```python

df = pd.read_csv("/content/1774404110.csv")




```

```python
print(df.head(10))
```

```python
print("Rows and Columns:", df.shape)
```

```python
print(df.describe())
```

2

```python
filtered_df = df[(df['model year'] == 75) & (df['weight'] < 3000)]

result = filtered_df[['car name', 'weight', 'mpg']]

print(result)
```

3

```python

print(df.isnull().sum())

```

```python
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')
```

```python
median_hp = df['horsepower'].median()
df['horsepower'].fillna(median_hp, inplace=True)
```

```python
print("Missing values after fix:")
print(df.isnull().sum())
```

4

```python
df['power_to_weight_ratio'] = df['horsepower'] / df['weight']

print(df[['horsepower', 'weight', 'power_to_weight_ratio']].head())
```

5

```python
grouped = df.groupby('origin')['mpg'].mean()

print(grouped)
```

6

```python

summary_df = df.groupby('model year')[['mpg', 'weight', 'horsepower']].mean()

summary_df = summary_df.reset_index()

print(summary_df)
```

7

```python

high_mpg_df = df[df['mpg'] > 30][['mpg', 'cylinders', 'horsepower', 'weight']]

high_mpg_df.to_csv('high_mpg_cars.csv', index=False)

print("File saved successfully!")
```
